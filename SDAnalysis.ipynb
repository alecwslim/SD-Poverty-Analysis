# Abstract

### Abstract

The primary motivation for this research project was to investigate the relationship between proximity to military and industrial land, poverty, and the rates of environmental illness in cities within San Diego County, California. Given the significant industrial and military presence in the area, particularly around the Port of San Diego, we hypothesized that both proximity to such areas and poverty would be strong predictors of environmental illness rates. Previous studies have highlighted the health risks associated with living near industrial sites in other cities, prompting us to explore if similar patterns exist in San Diego.

Our analysis revealed an unexpected result: Proximity to military and industrial land was not a statistically significant predictor of environmental illness rates, with a $p$-value of 0.117. This finding contrasts with existing literature that associates industrial proximity with increased health risks, suggesting that other factors may be mitigating this effect in San Diego County or that there exist socioeconomic factors leading individuals ill with non-fatal conditions in areas more affected by poverty to avoid seeking treatment (thereby skewing the resulting county-wide data).

In contrast, our findings confirmed that poverty is a significant predictor of environmental illness rates. The data showed a strong positive correlation between higher poverty rates and increased instances of environmental illnesses, underscoring the significant impact of socioeconomic status on health outcomes. This highlights the critical role that socioeconomic factors play in environmental health disparities, particularly in low-income cities such as National City.

In conclusion, while proximity to military and industrial areas may not be as influential as previously thought, addressing poverty is crucial for mitigating environmental health disparities in San Diego County. These results call for a comprehensive approach to public health that includes socioeconomic interventions and targeted policies to improve living conditions and healthcare access for impoverished communities, thereby promoting health equity across the region.

# Research Question

What is the relationship between a neighborhood's proximity to San Diego's port, heavy industry, or military bases, its poverty rate, and its rates of lung disease, cancer, and other environmental illnesses?

## Background and Prior Work

In many ways, the Port of San Diego is the city's backbone.  Opened in 1962, it is responsible for bringing everything from Honda automobiles to passengers aboard cruise ships into the city.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)  While the heavy industry and tourism enabled by the port contributed to transforming what was a small military town into America's eighth largest city within a generation<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2), it does not come without its own drawbacks.  De Roos et. al. have shown that residential areas near industrial sites in Los Angeles, Seattle, and Detroit are all at greater risk of health complications such as lymphoma, lung disease, and other environmental diseases.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)  However, they also showed that the rate of occurrence for these complications varies not only with proximity to an industrial site, but the type of goods produced at each industrial site as well.  In particular, individuals living within half a mile of industrial plants producing rubber, refining metal, or processing chemicals are over twice as likely to develop non-Hodgkin lymphoma versus individuals who do not.  In a similar vein, a 2023 report on water pollution from PFAS and other "forever chemicals" traced back to dumping on military bases suggests a similarly bleak picture for residents in affected areas being over twice as likely to develop kidney, testicular, or prostate cancer.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)  Ultimately, reports such as these lead to the same question:  Does the Port of San Diego and its nearby industrial plants bring about similar health concerns for San Diegans?

This question is a novel one because much of the discussion around environmental illness in San Diego revolves around sewage flow northward from Mexico.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5)  While this has been directly linked with a number of gastrointestinal complications particularly in the South San Diego Bay<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6), this means that other environmental hazards are comparatively under-researched.  There have been several high-profile cases -- fumigation of methyl bromide pesticide affecting liver and kidney health in Barrio Logan residents starting in the 1990s<a name="cite_ref-7"></a>[<sup>7</sup>](#cite_note-7), as well as diesel particulate matter causing air quality concerns in Barrio Logan and Logan Heights<a name="cite_ref-8"></a>[<sup>8</sup>](#cite_note-8) -- but a comprehensive understanding of the role of the port, heavy industry, and military bases on local communities' health is largely absent.  One factor that remains constant between existing analyses is the socioeconomic status and racial makeup of affected neighborhoods.  Residents of Barrio Logan, over 80% of whom are Latino, experience some of the highest rates of poverty in San Diego.<a name="cite_ref-9"></a>[<sup>9</sup>](#cite_note-9)  As a result, the lack of comprehensive data analysis on environmental illness in such neighborhoods is actively detrimental to health of its residents.

# Hypothesis


We predict that there will be a strong positive correlation between neighborhoods' proximity to the Port of San Diego (as well as Naval Base San Diego) and rates of lung disease, cancer, and other environmental illnesses.  These rates will be connected to high-profile environmental disruptions -- such as methyl bromide fumigation mentioned above -- but will be more readily explained by less egregious yet more consistent pollutants.  We expect to find this positive correlation between proximity and illness strongest among neighborhoods with higher rates of poverty, such as Barrio Logan, Logan Heights, and National City.  Conversely, we also expect to find lower rates of illness in wealthier neighborhoods even if they are in close proximity to the port and naval base (such as Gaslamp Quarter and Marina).

**Dataset #1:**

The most important variables in the San Diego County 2021 Non-Communicable (Chronic) Diseases dataset are "Geography" (region in San Diego), "CONDITION" (specific health conditions), "OUTCOME" (specific health outcomes, such as recovery or death), "IndexGroup" (demographic data), and "Rates" (rate of a specific condition). The rate of diseases are quantified by number per 100,000 people in that region and demographic group (e.g. race, gender, age). In order to clean up and preprocess this dataset, we have to remove missing data because many regions lack disease rates for particular demographic groups.

**Dataset #2:**

The most important variables in the Poverty Level by City dataset are "Geography" (the specific city name), "<50% FPL" (the percentage of residents whose household income is less than half of the federal poverty line), "50-74% FPL" (similarly to the previous column), and "75-99% FPL."  This percentages are written relative to the federal poverty line for each household when accounting for size, meaning relative to \\$15,060 per year for one-person households, \\$20,440 for two-person, \\$25,820 for three-person, \\$31,200 for four-person, up to \\$52,720 for eight-person.

**Dataset #3:**

The third dataset is an ESRI Shapefile containing all current land-use and zoning in place for San Diego County.  Its most important variable is "lu," the land use code as published on [SANDAG's website](https://www.sandag.org/-/media/SANDAG/Documents/PDF/regional-plan/sustainable-growth-and-development/land-use/land-use-codes-2022.pdf).  We will be using this dataset to determine a city's proximity to heavy industry (code 2001), marine terminals (4120), or military use land (6701).  Because the dataset's `.shp` is very large (443 MB), it is not able to be stored as part of our remote GitHub repository.  However, we will be able to extract the relevant regions' latitude and longitude in order to compute proximity to each of the cities being studied.

---

**Merging Datasets**:

Because the Poverty Level by City dataset has no missing or invalid data, our only preprocessing step is to remove the "City of" prefix from each row in the "Geography" column.  This allows us to connect this dataset to the Non-Communicable Diseases dataset by identifying matching "Geography" columns and summing the "<50% FPL," "50-74% FPL," and "75-99% FPL" columns.  This will yield a merged dataset that contains health outcome data and the percentage of residents in that area living under the federal poverty line.

Additionally, we can compute individual cities' latitude and longitude using [OpenStreetMap Nominatim](https://nominatim.org) via [GeoPy](https://geopy.readthedocs.io/en/stable) and store them in columns within the Disease data frame.  This will allow us to compute distances from areas of interest within the Land Use shapefile, without merging the two into a single data frame.

## San Diego County 2021 Non-Communicable (Chronic) Diseases

```python
import pandas as pd
import numpy as np
import geopandas as gpd
from geopy.geocoders import Nominatim

# Read CSV
df_disease_raw = pd.read_csv('./datasets/2021_Non-Communicable__Chronic__Diseases_20240513.csv')
df_disease = df_disease_raw.dropna().reset_index(drop=True)

# Compute latitude and longitude using Nominatim
geolocator = Nominatim(user_agent="ucsd_cogs108")

tried_cities = {}
lat_long = {}
for city in df_disease['Geography'].values:
    if city in tried_cities or city in lat_long:
        continue
        
    tried_cities[city] = True
    location = geolocator.geocode(city + ', California, USA')
    if location is None:
        continue
    
    lat_long[city] = (location.latitude, location.longitude)

# Add lat/long as columns to data
df_disease['Latitude'] = df_disease['Geography'].apply(lambda city: lat_long[city][0] if city in lat_long else np.nan)
df_disease['Longitude'] = df_disease['Geography'].apply(lambda city: lat_long[city][1] if city in lat_long else np.nan)

df_disease.head()
```

## 2016 San Diego County Demographics - Poverty Level by City

```python
df_poverty = pd.read_csv('./datasets/2016_San_Diego_County_Demographics_-_Poverty_Level_by_City_20240516.csv')
# Remove "City of" prefix
df_poverty['Geography'] = df_poverty['Geography'].str.replace('City of ', '')
# Compute total poverty rate statistic
df_poverty['PovertyRate'] = df_poverty['<50% FPL'] + df_poverty['50-74% FPL'] + df_poverty['75-99% FPL']

# Merge with df_disease dataset
df_disease['PovertyRate'] = 0
rates = {}
for index, row in df_poverty.iterrows():
    rates[row['Geography']] = row['PovertyRate']

df_disease['PovertyRate'] = df_disease.apply(
    lambda x: rates[x['Geography']] if x['Geography'] in rates else np.nan,
    axis=1
)
df_disease = df_disease.dropna()

df_disease.head()
```

## Exploratory Data Analysis

Given our three datasets, there are a few questions to be asked before we can connect disease, poverty, and land use data together.  With respect to the environmental illnesses themselves, we wish to find their particular distributions.  We also wish to explore how a city's poverty rate affects its rate of disease.

For the land use dataset, we must understand the distribution of military bases or heavy industry within San Diego County:  Are they clustered anywhere in particular, and do they tend more towards the ocean or more inland?  Along with this, we wish to understand the differences between cities nearby these land use zones and those far away.

### Section 1:  Disease Distributions

First, let us explore the contents of the disease dataset.  We will first strip any unnecessary or redundant fields, limit our analysis to known environmental illnesses, and explore the data at the city level.

```python
import seaborn as sns
from matplotlib import pyplot as plt

sns.set_theme(rc={'figure.figsize':(14,8)})
sns.set_style('darkgrid')
plt.xticks(rotation=45)

# Remove unnecessary fields (e.g. Year, ConditionGroup, etc.)
df_disease_stripped = df_disease[['Geography', 'CONDITION', 'OUTCOME', 'Index', 'Rates', 'Latitude', 'Longitude', 'PovertyRate']]

# Include only total rates (i.e. exclude data for specific demographic groups)
df_disease_total = df_disease_stripped[df_disease_stripped['Index'] == 'Total']

sns.stripplot(x='Geography', y='Rates', hue='CONDITION', data=df_disease_total).set_title('City vs. Disease Rates')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3)
```

Each city appears to follow a relatively similar distribution:  One disease is much more common than the others, and many diseases are relatively uncommon.  Let us now explore what specific diseases these are.

```python
# Filter dataset to deaths and group by condition
df_condition_death = df_disease_total[df_disease_total['OUTCOME'] == 'Death']\
    .groupby('CONDITION', as_index=True)\
    .first()[['Rates']]\
    .sort_values('Rates', ascending=False)\
    .reset_index()

plt.xticks(rotation=90)
sns.barplot(x='CONDITION', y='Rates', data=df_condition_death).set_title('Condition vs. Rate of Death')
```

We see that overall cancer and overall heart disease dominate the list, with congential heart disease (CHD) and stroke following after.  Let us perform the same analysis for non-fatal outcomes.

```python
# Filter dataset to non-deaths and group by condition
df_condition_death = df_disease_total[df_disease_total['OUTCOME'] != 'Death']\
    .groupby('CONDITION', as_index=True)\
    .first()[['Rates']]\
    .sort_values('Rates', ascending=False)\
    .reset_index()

plt.xticks(rotation=90)
sns.barplot(x='CONDITION', y='Rates', data=df_condition_death).set_title('Condition vs. Non-Fatal Case Rate')
```

Interestingly, heart disease is exceedingly common (and hypertension follows).  We also notice that some diseases which did not result in death (e.g. Lupus) have non-fatal cases included in the dataset.  Also, rates of overall cancer are significantly lower, suggesting that cancer is a highly fatal condition.  Next, let us explore how poverty rates affect these distributions at a city level.

```python
# Group disease dataset by city name
df_geog_poverty = df_disease_total[['Geography', 'PovertyRate']].groupby(by='Geography').first().sort_values('PovertyRate', ascending=False)

poverty_rate = df_geog_poverty.to_dict()['PovertyRate']

plt.xticks(rotation=45)
sns.barplot(x='Geography', y='PovertyRate', data=df_geog_poverty.reset_index()).set_title('City vs. Poverty Rate')
```

We see that National City has the highest rate of poverty at 24.2%, while Coronado has the lowest rate at 5.8%.  We will compare disease rates in these two cities to determine if there is a meaningful difference between their disease distributions.

```python
df_national_city = df_disease_total[df_disease_total['Geography'] == 'National City']\
    .groupby('CONDITION', as_index=True)\
    .first()[['Rates']]\
    .sort_values('Rates', ascending=False)\
    .reset_index()

df_coronado = df_disease_total[df_disease_total['Geography'] == 'Coronado']\
    .groupby('CONDITION', as_index=True)\
    .first()[['Rates']]\
    .sort_values('Rates', ascending=False)\
    .reset_index()

fig, axs = plt.subplots(ncols=2)
axs[0].tick_params(labelrotation=90)
axs[0].set_ylim([0, 300])
axs[1].tick_params(labelrotation=90)
axs[1].set_ylim([0, 300])
sns.barplot(x='CONDITION', y='Rates', data=df_national_city, ax=axs[0]).set_title('Condition vs. Disease Rate (National City)')
sns.barplot(x='CONDITION', y='Rates', data=df_coronado, ax=axs[1]).set_title('Condition vs. Disease Rate (Coronado)')
```

The shape of these distributions is quite similar, however it is worth noting that asthma is the dominant condition in National City while hypertension dominates in Coronado.  Beyond this, rates of cancer and heart disease appear approximately the same between the two cities (around 150 cases per 100,000 people).  Let us also perform this analysis for every city in the dataset, first color coding each disease.

```python
fig, axs = plt.subplots(5, 3, figsize=(14, 14))
fig.suptitle('City vs. Disease Rate')

# Generate consistent color scheme
colors = sns.color_palette('hls', len(df_disease_total['CONDITION'].unique()))
palette = dict(zip(df_disease_total['CONDITION'].unique(), colors))

# Draw plot for each city
i = 0
j = 0
for city in tried_cities:
    df_city = df_disease_total[df_disease_total['Geography'] == city]\
        .groupby('CONDITION', as_index=True)\
        .first()[['Rates']]\
        .sort_values('Rates', ascending=False)\
        .reset_index()
    
    if len(df_city) == 0:
        continue
    
    title = city + ' (' + '{:.1f}'.format(poverty_rate[city] * 100) + '% Poverty Rate)'
    axs[i][j].get_xaxis().set_visible(False)
    axs[i][j].set_ylim([0, 300])
    sns.barplot(x='CONDITION', y='Rates', data=df_city, ax=axs[i][j], palette=palette).set_title(title)
    
    i += 1
    if i > 4:
        i = 0
        j += 1

# Add legend subplot
legend_ax = axs[4][2]
legend_ax.set_axis_on()
legend_ax.axis('off')

# Create legend handles
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=palette[cond], markersize=10, label=cond) for cond in df_disease_total['CONDITION'].unique()]
legend_ax.legend(handles=handles, loc='center')

# Delete unused plots
fig.delaxes(axs[3][2])
```

A few relationships and properties reveal themselves from these charts:
- Asthma is the most common condition among residents of National City, El Cajon, Lemon Grove, and Chula Vista.  All four of these cities have a poverty rate of 12% or higher, with the former two being above 24%.  These rates of asthma are the only outliers apparent in the data, and therefore may be worth exploring in more detail.
- In cities with lower poverty rates (i.e. less than 12%), cancer and heart disease are the most common conditions by a wide margin.
- Rates of cancer and heart disease appear to be relatively consistent between the cities, given that each subplot has the same scale.

Therefore we observe a variety of differences in rates of disease between various cities, rates of fatal versus non-fatal diseases, and the possible impact of poverty rates on illness.  However, we have yet to process our San Diego County land use data in conjunction with our disease data.

### Section 2:  Land Use

We now wish to connect our disease dataset to the San Diego County land use Shapefile.  First let us filter out only plots of land that are relevant to our analysis, namely military and heavy industry plots, in order to compute each city's proximity to its closest plot.

```python
land_use_codes = [
    1403, # Military barracks
    2001, # Heavy industry
    2201, # Extractive industry (e.g. oil drilling)
    4102, # Military airport
    6701, # Military use
    6702  # Military training
]

# Filter only relevant land plots
gdf_industry_military = gdf_landuse[gdf_landuse['lu'].isin(land_use_codes)].reset_index()

# Rename lu column to friendly names for plotting
lu_to_string = {
    1403: 'Military barracks',
    2001: 'Heavy industry',
    2201: 'Extractive industry',
    4102: 'Military airport',
    6701: 'Military use',
    6702: 'Military training'
}
gdf_industry_military_friendly = gdf_industry_military.copy()
gdf_industry_military_friendly['lu'] = gdf_industry_military_friendly['lu'].apply(lambda lu: lu_to_string[lu])

fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_landuse.plot(ax=ax)
gdf_industry_military_friendly.plot(ax=ax, column='lu', legend=True, categorical=True, cmap='Reds')
ax.set_title('Military and Industrial Land Use')
```

Relevant portions of land are highlighted in red.  Here, we see that there are some regions closer to the City of San Diego (e.g. Point Loma).  We also notice that the majority of included regions are far in North County, likely making up Camp Pendleton in Oceanside.  Let us now compute centroids for each plot of land, in order to allow for comparison against our disease dataset.

```python
gdf_centroids = gdf_industry_military.copy()

# Update geometry to use centroid instead of polygon
gdf_centroids.geometry = gdf_centroids['geometry'].centroid

# Reproject with new CRS (latitude/longitude, EPSG 4326)
gdf_centroids = gdf_centroids.to_crs(epsg=4326)

# Rename lu column to friendly names for plotting
gdf_centroids_friendly_lu = gdf_centroids.copy()
gdf_centroids_friendly_lu['lu'] = gdf_centroids_friendly_lu['lu'].apply(lambda lu: lu_to_string[lu])

# Plot latitude/longitude points
fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_centroids_friendly_lu.plot(ax=ax, column='lu', legend=True, categorical=True, cmap='Blues').set_title('Military and Industrial Land Use')
```

While this plot looks similar to the previous, we notice that Camp Pendleton is significantly closer to downtown than before.  This is due to the reprojection from San Diego County's surveying coordinate system into latitude and longitude compressing the map vertically.  This will not affect our analysis, but it demonstrates that latitude and longitude (like any map or Cartesian coordinate system) are an approximation of reality.

Let us now overlay each city onto this map.

```python
# Get first entry with given city name
df_cities = df_disease_total.groupby('Geography').first()

# Construct GeoDataFrame from DataFrame
gdf_disease = gpd.GeoDataFrame(df_cities, geometry=gpd.points_from_xy(df_cities['Longitude'], df_cities['Latitude']), crs='epsg:4326').reset_index()

# Plot points
fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_centroids_friendly_lu.plot(ax=ax, column='lu', legend=True, categorical=True, cmap='Blues')
gdf_disease.plot(ax=ax, markersize=200, column='Geography', legend=True, categorical=True, cmap='autumn').set_title('Cities')
```

Interestingly, we see that the only cities in very close proximity to industry or military land are those closer to Downtown.  Cities towards the north (such as Carlsbad or Oceanside) are adjacent to Camp Pendleton, but not in such close proximity as with National City or Chula Vista.

Let us now repeat this plot, but factor in poverty rate for each city's hue.

```python
# Plot points
fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_centroids_friendly_lu.plot(ax=ax, column='lu', legend=True, categorical=True, cmap='Blues')
gdf_disease.plot(ax=ax, markersize=200, column='PovertyRate', legend=True, legend_kwds={'label': 'Poverty Rate (Fraction)'}, cmap='Reds').set_title('Cities by Poverty Rate')
```

We notice a few interesting properties of this map:
- North County cities (e.g. Oceanside and Carlsbad) do not have exceptionally high poverty rates, and are relatively far from relevant plots of land.
- A few cities such as National City and El Cajon are very close to relevant plots, and also have high poverty rates.  However, Coronado is also close but has a very low poverty rate.

Let us now compute each city's distance from its closest military/industry plot of land and factor this into our map.

```python
# Compute distance between city centroid and closest land use centroid
def get_distance(row):    
    out = None
    min_dist = float('inf')
    for (_, point) in gdf_centroids.iterrows():
        distance = row.distance(point['geometry'])
        if distance < min_dist:
            min_dist = distance
            out = point
    
    return min_dist

gdf_disease['Distance'] = gdf_disease['geometry'].apply(get_distance)

# Plot points
fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_centroids_friendly_lu.plot(ax=ax, column='lu', legend=True, categorical=True, cmap='Blues')
gdf_disease.plot(ax=ax, markersize=pow(100*(0.1-gdf_disease['Distance']), 3), column='PovertyRate', legend=True, legend_kwds={'label': 'Poverty Rate (Fraction)'}, cmap='Reds').set_title('Cities by Proximity and Poverty Rate')
```

As we suspected, National City, Chula Vista, and Coronado are among the largest markers on the map (implying they are closest to a relevant plot of land).  However, Oceanside and El Cajon are close as well, while all others (Lemon Grove, La Mesa, Poway, etc.) are significantly smaller. ## Analysis

Let us now determine if the conclusions suggested by our EDA are supported by statistical analyses.  Because of the geospatial nature of our data, we apply two geospatial analyses:  Hypothesis testing with Monte Carlo simulation, and covariate testing with the Poisson Point Process model.

### Section 1:  Monte Carlo Hypothesis Testing

We wish to test our hypothesis that there is a positive correlation between proximity to military/industrial land, poverty rate, and rates of environmental illnesses.  To do this, we will apply Monte Carlo simulation and compute $p$-values using average nearest neighbor analysis.  This approach is appropriate because it makes no assumptions about the distribution of our dataset (allowing us to perform hypothesis testing on data which may or may not be normally distributed) *and* does not suffer from many of the issues that other spatial analyses do, such as the modifiable areal unit problem or edge effects.  However, it fails to properly account for covariate effects as will be detailed in the following section.

To perform Monte Carlo Simulation, we can first formulate our null hypothesis -- namely that rates of environmental illness are consistent with complete spatial randomness (CSR).  Then, we can simulate this random process and compare its measured average nearest neighbors (ANN) to that of our original data.

```python
from scipy.spatial import cKDTree

# Helper function for performing ANN testing
def compute_ann(df):
    # Compute composite column with disease and poverty rates
    df_copy = df.copy()
    df_copy['CompositeRate'] = df_copy['Rates'] * df_copy['PovertyRate']
    
    # Generate GeoDataFrame from dataset
    gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df['Longitude'], df['Latitude']), crs='epsg:4326').reset_index()
    
    # Convert DataFrame to NumPy array
    coords = np.array([(geom.x, geom.y) for geom in gdf.geometry])
    weights = df_copy['CompositeRate'].values
    
    # Create k-d tree for fast lookups
    tree = cKDTree(coords)
    
    # Query the tree for the nearest neighbor distances
    distances, _ = tree.query(coords, k=2)
    nearest_distances = distances[:, 1]
    weighted_distances = nearest_distances * weights
    weighted_observed_mean_distance = np.sum(weighted_distances) / np.sum(weights)
    
    # Calculate weighted expected mean nearest neighbor distance
    total_area = (gdf.total_bounds[2] - gdf.total_bounds[0]) * (gdf.total_bounds[3] - gdf.total_bounds[1])
    weighted_density = np.sum(weights) / total_area
    weighted_expected_mean_distance = 1 / (2 * np.sqrt(weighted_density))
    
    # Compute the resulting ANN
    return weighted_observed_mean_distance / weighted_expected_mean_distance

# Reduce DataFrame from one row per disease to one row per city with a summed disease rate
df_environmental = df_disease_total.groupby(['Geography', 'Latitude', 'Longitude', 'PovertyRate'])['Rates'].sum().reset_index()
df_environmental.columns = ['Geography', 'Latitude', 'Longitude', 'PovertyRate', 'Rates']

# Compute ANN
observed_ann = compute_ann(df_environmental)
print('ANN for observed dataset:\t' + str(observed_ann))
```

In the above cell, we define a helper function for computing an input dataset's average nearest neighbors, or the ratio between a data point's mean distance to its nearest neighbor and the expected mean distance to its nearest neighbor.  Then, we weight these values by a newly-created column `CompositeRate` to compute our final weighted ANN value.  Notice that the value being used is disease rate *scaled by poverty rate*, attempting to encapsulate the role of both variables.

Now, let us introduce the randomized process for redistributing data points.

```python
# Load simplified landuse generated offline with:
# gdf_landuse['new_column'] = 0
# gdf_landuse.dissolve(by='new_column').simplify(tolerance=100).to_file('landuse_simple.shp')
gdf_landuse_simple = gpd.read_file('./datasets/landuse_simple/landuse_simple.shp')

def randomize_data(df):
    df_copy = df.copy()
    
    # Sample uniformly over the entirety of San Diego County
    sampled_points = gdf_landuse_simple.sample_points(len(df_copy))
    
    # Convert these sampled points from California Zone 6 coordinates to latitude/longitude
    sampled_latlong = sampled_points.to_crs(epsg=4326)
    
    # Explode multipoint into many single points
    sampled_exploded = sampled_latlong.explode(index_parts=True)[0]
    
    # Modify output DataFrame with new lat/long
    idx = -1
    def get_y(row):
        nonlocal idx
        idx += 1
        return sampled_exploded[idx].y
    df_copy['Latitude'] = df_copy['Latitude'].apply(get_y)
    
    idx = -1
    def get_x(row):
        nonlocal idx
        idx += 1
        return sampled_exploded[idx].x
    df_copy['Longitude'] = df_copy['Longitude'].apply(get_x)
    
    return df_copy
    
df_random = randomize_data(df_environmental)
gdf_random = gpd.GeoDataFrame(df_random, geometry=gpd.points_from_xy(df_random['Longitude'], df_random['Latitude']), crs='epsg:4326').reset_index()

fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_landuse_simple.to_crs(epsg=4326).plot(ax=ax)
gdf_random.plot(color='orange', markersize=100, ax=ax)
ax.set_title('Randomized City Locations')
```

The `randomize_data` function redistributes each of the data points in the input to a random position within San Diego County, making the dataset consistent with CSR (i.e. uniformly distributed within the County).  Now, let us compute average nearest neighbors for 1000 randomized runs and perform hypothesis testing using this new data.

```python
simulated_anns = []
for i in range(0, 1000):
    simulated_anns.append(compute_ann(randomize_data(df_environmental)))
fig, ax = plt.subplots(1, 1, figsize=(14, 8))
sns.histplot(simulated_anns, ax=ax)
ax.set(xlabel='Computed Average Nearest Neighbors Ratio', ylabel='Count')
ax.set_title('Average Nearest Neighbors Ratio (Randomized Locations)')
```

This histogram is the computed ANN statistic for each of 1000 randomized simulations.  Clearly, it is relatively normally distributed as would be expected with uniformly-distributed data.

Now, we compute the $p$-value using the following formula described at the [spatial analysis resource referenced by COGS 108 lecture materials](https://mgimond.github.io/Spatial/hypothesis-testing.html) to test our hypothesis:

$ p = \frac{min(N_{greater} + 1, N + 1 - N_{greater})}{N + 1} $

```python
# N_greater is the number of values for which the simulated statistic is greater than the test statistic
N_greater = len([ann for ann in simulated_anns if ann > observed_ann])

# N is the number of observations
N = len(simulated_anns)

# Compute p value using provided formula
p = min(N_greater + 1, N + 1 - N_greater) / (N + 1)
print('Final p value:\t' + str(p))
print('p < 0.05?\t' + str(p < 0.05))
```

**Thus, $p$ is less than 0.05 therefore we reject the null hypothesis**.  This result is unsurprising given the qualitative difference explored during EDA, and our value suggests that there is a fewer than 2% chance the relationship between proximity to industrial/military land, poverty, and environmental illness observed in the data came about as a result of a random process.

However, the composite test statistic `CompositeRate` likely introduces bias and does not fully consider poverty rate as a covariate.  We wish to address this in the following section.

### Section 2:  Poisson Point Process

While Monte Carlo simulation suggests that we reject the null hypothesis, poverty rate is insufficiently accounted for in ANN calculation.  Therefore, we can apply a Poisson Point Process which better accounts for the role of poverty as a covariate.  To do this, we apply a Generalized Linear Model (GLM) to fit a Poisson regression model with poverty rate and proximity as predictors:

```python
import statsmodels.api as sm
from statsmodels.formula.api import glm
from statsmodels.genmod.families import Poisson

# Convert environmental illness data into a GeoDataFrame
gdf_environmental = gpd.GeoDataFrame(df_environmental, geometry=gpd.points_from_xy(df_environmental['Longitude'], df_environmental['Latitude']), crs='epsg:4326').reset_index()
gdf_environmental['Distance'] = gdf_environmental['geometry'].apply(get_distance)

# Convert the spatial coordinates to a suitable format
gdf_environmental['x'] = gdf_environmental.geometry.x
gdf_environmental['y'] = gdf_environmental.geometry.y

# Prepare the data for the Poisson regression model
formula = 'Rates ~ PovertyRate + Distance'
model = glm(formula=formula, data=gdf_environmental, family=Poisson()).fit()

# Print the summary of the model to analyze the effect of the covariate
print(model.summary())
```

This analysis suggests something different from the prior section, namely that proximity is much less statistically significant.  In fact, it states that poverty rate is correlated much more closely (with $p=0.000$) while proximity has more than a 10% probability of having been sampled from a random distribution (with $p=0.117$).  **Thus, the Poisson Point Model suggests that poverty is the more telling predictor of a city's rates of environmental illness, with proximity to military or industrial land being only a minor predictor.**

Let us now visualize the model's predictions for each city in the dataset, given its poverty rate and computed proximity to military/industrial land:

```python
# Predict the intensity of the point process
gdf_environmental['Intensity'] = model.predict(gdf_environmental[['PovertyRate', 'Distance']])

# Visualize the predicted intensity if needed
fig, ax = plt.subplots(1, 1, figsize=(14, 8))
gdf_landuse_simple.to_crs(epsg=4326).plot(ax=ax)
gdf_environmental.plot(column='Intensity', ax=ax, markersize=150, legend=True, cmap='Reds', 
         legend_kwds={'label': 'Cases per 100,000'})
plt.title('Predicted Intensity of Environmental Illness Cases')
plt.show()
```

Clearly, this map is similar to the asthma map generated in the EDA section:  National City and El Cajon have elevated rates, while all others have normal to low rates.  However, the conclusion suggested by this model is that poverty is the primary predictor of such a result.
